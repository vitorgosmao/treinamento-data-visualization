## Fazendo primeiras visualizações

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import datetime
```

```python
df = pd.read_csv('base de dados/monitoramento_tempo.csv')
```

```python
df.info()
```

```python
df['data'] = pd.to_datetime(df['data'])
```

```python
df['data']
```

```python
plt.plot(df['data'], df['temperatura'])
```

```python
plt.figure(figsize=(15,8))
plt.plot(df['data'], df['temperatura'])
plt.title('Temperatura no tempo', fontsize=16)
```

```python
fig = plt.figure(figsize= (15,8))# objeto figura
eixo = fig.add_axes([0,0,1,1])# adicionar os eixos da figura(x,y)

eixo.plot(df['data'], df['temperatura'], color= 'green' )# inserindo uma visualização dentro da figura

eixo.set_title('Temperatura no momento', fontsize = 25)# definindo um titulo para o gráfico

eixo.set_ylabel('Temperatura', fontsize = 20)# definindo a legenda/label do eixo y
eixo.set_xlabel('Data',  fontsize = 20)# definindo a legenda/label do eixo x

eixo.legend(['temperatura'], loc = 'lower right', fontsize = 15)# legenda do que está dentro do gráfico

```

```python
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias]
temperatura_por_dia_da_semana
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
indice = range(len(temperatura_por_dia_da_semana))

ccores = ['black', 'r', 'g', 'b', 'yellow', 'orange', 'magenta']

eixo.bar(indice, temperatura_por_dia_da_semana, color=cores, edgecolor = 'k')
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura média', fontsize=15)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index,
        autopct='%.1f%%')
eixo.set_title('Temperatura por dia da semana', size=15, pad=10)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [0.1,0,0,0,0,0.1,0.1]

eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index,
        autopct='%.1f%%', explode=explodir, shadow=True)
eixo.set_title('Temperatura por dia da semana', size=15, pad=10)
```
